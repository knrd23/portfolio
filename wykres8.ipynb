```python
f=open("wykres7.txt","r", encoding='utf-8')
tekst=f.read().split()
f.close()
```

```python
slowa1=list(tekst)
```

```python
slowa2=list(map(lambda x: x.replace(',',''), slowa1))
```

```python
znak=["?",".","--",":","(",")","!",";","—",'"',"'",'—','…','...']
for znaki in znak:
  slowa2=list(map(lambda x: x.replace(znaki,''), slowa2))
```

```python
slowa3=[it for it in slowa2 if len(it)>=5]
```

```python
slowa4=list(map(lambda x: x.lower(), slowa3))
```

```python
from collections import Counter
slowa5=Counter(slowa4)
slowa6=list(slowa5.most_common(150))
slowa6
```

```python
doWykresu=dict(slowa6)
doWykresu
```

```python

import numpy as np
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import wordcloud as wr
```

```python
sprout_mask=np.array(Image.open("sprout.png"))
```

```python
wordcloud=wr.WordCloud(collocations=False,background_color='white',width=3000,height=2000,mask=sprout_mask).generate_from_frequencies(doWykresu)
```

```python
import matplotlib.pyplot as plt
image_colors = ImageColorGenerator(sprout_mask)
fig,ax=plt.subplots(figsize=(12.5,6.5))
ax.imshow(wordcloud,interpolation='bilinear')
ax.axis('off')
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.title('150 najczęściej występujących słów w "Chłopach" Reymonta długości co najmniej 5',
          fontdict={'fontsize':8},
          bbox=dict(facecolor='green',alpha=0.5))
plt.savefig('wykres8.pdf',dpi=500)
plt.show()
```
