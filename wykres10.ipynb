```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('wykres10.csv',delimiter=';')
df
```

```python
df1 = df.melt(id_vars='Produkty', var_name='Rok', value_name='Cena')
df1
```

```python
df1['Cena'] = df1['Cena'].str.replace(',', '.')
df1['Cena'] = df1['Cena'].astype(float)
```

```python
df1
```

```python
import seaborn as sns
```

```python
kolor=['cadetblue', 'teal', 'steelblue', 'powderblue','darkturquoise','lightskyblue']
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Produkty', y='Cena', data=df1, palette=kolor)
sns.set_style("darkgrid")
plt.title('Ceny skupu mięsa zwierząt hodowlanych w latach 2005-2023 za 1 kg [zł]',fontsize=18,
          bbox={'color':'green','alpha':0.5},
          pad=15)
plt.xlabel('Mięso zwierząt hodowlanych')
plt.ylabel('Cena za 1 kg w wadze żywej [zł]')
nazwy=['bydło (bez cieląt)','cielęta','trzoda chlewna','owce','konie','drób']
plt.xticks(range(len(nazwy)),nazwy,rotation=45)
plt.tight_layout()
plt.savefig('wykres10.pdf',dpi=300)
plt.show()
```
