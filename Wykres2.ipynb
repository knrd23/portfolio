```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gdp
```

```python
df2=pd.read_csv('wykres2.csv',sep=';')
df2.drop(0,inplace=True)
df2
```

```python
mapa_woj = gdp.read_file('wojewodztwa.geojson')
mapa_woj['JPT_NAZWA_']=['slaskie','opolskie','wielkopolskie','zachodniopomorskie','swietokrzyskie','kujawsko-pomorskie',
                        'podlaskie','dolnoslaskie','podkarpackie','malopolskie','pomorskie','warminsko-mazurskie','lodzkie',
                        'mazowieckie','lubelskie','lubuskie']
mapa_woj.sort_values(by='JPT_NAZWA_',inplace=True)
mapa_woj
```

```python
df2=pd.read_csv('wykres2.csv',sep=';')
df2=df2.drop(index=0)
df2['Województwo']=['dolnoslaskie','kujawsko-pomorskie','lubelskie','lubuskie','lodzkie','malopolskie','mazowieckie','opolskie','podkarpackie','podlaskie','pomorskie','slaskie','swietokrzyskie','warminsko-mazurskie','wielkopolskie','zachodniopomorskie']
df2
```

```python
dowykresu=pd.merge(mapa_woj,df2,how='left',left_on='JPT_NAZWA_',right_on='Województwo')
dowykresu
```

```python
dowykresu=dowykresu[['geometry','Województwo','Ogółem']]
```

```python
dowykresu['Ogółem']=dowykresu['Ogółem'].apply(lambda x: x.replace(',','.'))
```

```python
dowykresu['Ogółem']=dowykresu['Ogółem'].apply(lambda x: float(x))
```

```python
dowykresu
```

```python

fig, ax = plt.subplots(figsize=(16, 9))

dowykresu.to_crs(epsg=2180).plot( #to_crs: metoda w gpd, przekształca do układu EPSG:2180 oparty na rzutowaniu Transverse Mercator
    ax=ax,
    column='Ogółem',
    cmap='Greens',
    linewidth=0.6,
    edgecolor='black',
    legend=True
)


ax.set_title('Średnia powierzchnia użytków rolnych w gospodarstwach rolnych\n według województw w 2020 roku [ha] ',
             size=20,
             bbox=dict(facecolor='green',alpha=0.5))


plt.tight_layout()
ax.set_axis_off()
plt.savefig('wykres2.pdf', dpi=300)
plt.tight_layout()
plt.show()
```
