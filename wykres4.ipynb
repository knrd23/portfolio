```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
dane=pd.read_csv('wykres4.csv',delimiter=';',names=['kod','województwo','łąki_i_pastwiska','bydło','mleko'])
dane
```

```python
dane.drop(index=0,inplace=True)
dane
```

```python
dane.drop(columns=['kod'],inplace=True)
dane
```

```python
dane['łąki_i_pastwiska']=dane['łąki_i_pastwiska'].str.replace(',','.')
dane['łąki_i_pastwiska']=dane['łąki_i_pastwiska'].str.replace(' ','')
dane['łąki_i_pastwiska']=dane['łąki_i_pastwiska'].astype(int)
```

```python
dane['bydło']=dane['bydło'].str.replace(',','.')
dane['bydło']=dane['bydło'].str.replace(' ','')
dane['bydło']=dane['bydło'].astype(int)
```

```python
dane['mleko']=dane['mleko'].str.replace(',','.')
dane['mleko']=dane['mleko'].str.replace(' ','')
dane['mleko']=dane['mleko'].astype(int)
```

```python
dane['mleko/bydło']=dane['mleko']/dane['bydło']
dane
```

```python
kolory_wojewodztw=dane['mleko/bydło']
kolory_wojewodztw

```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
x=dane['łąki_i_pastwiska']
y=dane['bydło']
size=dane['mleko']/100
labels=dane['województwo']
```

```python

fig=go.Figure(data=[go.Scatter(
    x=x,
    y=y,
    mode='markers',
    marker=dict(
        sizemode='area',
        size=size,
        color=kolory_wojewodztw,
        colorscale='RdPu',
        showscale=True,
        opacity=0.8,
        colorbar=dict(
            title='Ilość mleka przypadająca na sztukę bydła [tys. l/szt.]',
            title_font_size=16,
            title_side='right'
        )
    ),
    text=labels,
    hovertemplate='<b>%{text}</b><br>Łąki i pastwiska: %{x} ha<br>Bydło: %{y} szt.<br>Produkcja mleka: %{marker.size:.0f} tys. l <extra></extra>'
#                  nazwa województwa pogrubiona, koordynaty (x,y), rozmiar markera, dodatkowe informacje (tu puste)
)])
fig.update_layout(title_text='Pogłowie bydła [szt.] vs powierzchnia łąk i pastwisk trwałych [ha] <br> vs produkcja mleka [tys. l] vs wydajność bydła [tys. l/szt.] w roku 2020 ',
                  title_font_size=20,
                  title_x=0.5,
                  xaxis_title='Powierzchnia łąk i pastwisk trwałych [ha]',
                  yaxis_title='Pogłowie bydła [szt.]',
                  annotations=[
        dict(
            x=0.5,
            y=1.05,
            xref='paper',
            yref='paper',
            text='Rozmiar bąbelka oznacza wolumen produkcji mleka',
            showarrow=False,
            font=dict(size=12, color='#48488a')
        )
    ]
                  )
fig.write_html('wykres4.html')
fig.show()
```
